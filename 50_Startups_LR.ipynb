# Startup profit prediction
This particular dataset holds data from 50 startups in New York, California, and Florida. The features in this dataset are R&D spending, Administration Spending, Marketing Spending, and location features, while the target variable is: Profit.

# Domain Analysis

1. R&D Spend :-The amount which startups are spending on Research and development.
2. Administration :- The amount which startups are spending on the Admin panel.
3. Marketing Spend :- The amount which startups are spending on marketing strategies.
4. State :- To which state that particular startup belongs.
5. Profit :- How much profit that particular startup is making.

# Model Creation

```python
data.head()
```

```python
# Step1: Independent and dependent variable creation

X = data[["R&D Spend","Marketing Spend"]]
y = data.Profit  
```

```python
# Step 2: Training and testing data creation

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
# Step 3: Creating Model

from sklearn.linear_model import LinearRegression
LR=LinearRegression()    # Object creation
LR.fit(X_train,y_train)  # training the linear regression
y_predict=LR.predict(X_test)
```

```python
y_test       # actual values
```

```python
y_predict     # predicted values
```

```python
# step 4: Model Evoluation

from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
r2score=r2_score(y_test,y_predict)
```
